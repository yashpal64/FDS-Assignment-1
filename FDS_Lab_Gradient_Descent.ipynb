```python
#designing a gradient descent function
import numpy as np

def gradient_descent(x, y, learning_rate=0.01, iterations=1000):
    # Initialize parameters
    m = 0  # slope
    b = 0  # intercept
    n = len(y)  # number of data points

    # Perform gradient descent
    for _ in range(iterations):
        # Calculate predictions
        y_pred = m * x + b

        # Calculate gradients
        dm = (-2/n) * np.sum(x * (y - y_pred))  # gradient for slope
        db = (-2/n) * np.sum(y - y_pred)        # gradient for intercept

        # Update parameters
        m -= learning_rate * dm
        b -= learning_rate * db

    return m, b

# Example usage with generated data
x_values = np.random.uniform(-10, 10, 30)
y_values = 3 * x_values + 0.7 + np.random.normal(0, 1, 30)  # adding some noise

# Run gradient descent
learning_rate = 0.01
iterations = 1000
slope, intercept = gradient_descent(x_values, y_values, learning_rate, iterations)

print(f"Estimated slope: {slope:.4f}, Estimated intercept: {intercept:.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random x and y values
x_values = np.random.uniform(-10, 10, 30)
y_values = 3 * x_values + 0.7 + np.random.normal(0, 1, 30)  # Adding some noise

# Initialize parameters
w_old = np.random.uniform(-10, 10)
b_old = np.random.uniform(-10, 10)
print(f"Initial w: {w_old}, b: {b_old}")

# Calculate initial prediction and loss
y_cap = w_old * x_values + b_old
Loss = 1/2 * np.mean((y_cap - y_values)**2)

# Gradient descent
for i in range(500):
    # Calculate new parameters
    w_new = w_old - 0.01 * np.mean((y_cap - y_values) * x_values)
    b_new = b_old - 0.01 * np.mean((y_cap - y_values))

    # Update predictions and loss
    y_cap = w_new * x_values + b_new
    Loss = 1/2 * np.mean((y_cap - y_values)**2)

    # Update old parameters
    w_old = w_new
    b_old = b_new

    # Print loss every 10 iterations
    if i % 10 == 0:
        print(f"Iteration {i}: Loss = {Loss}")

print(f"Final w: {w_new}, Final b: {b_new}")

# Plotting the results
plt.scatter(x_values, y_values, color='blue', label='Data Points')
plt.plot(x_values, y_cap, color='orange', label='Fitted Line', linewidth=2)
plt.title('Gradient Descent Fit')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black', linewidth=0.5, ls='--')
plt.axvline(0, color='black', linewidth=0.5, ls='--')
plt.grid()
plt.legend()
plt.show()
#iteration vs epochs
```
